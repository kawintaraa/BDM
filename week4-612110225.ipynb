```python
df = pd.read_csv('https://raw.githubusercontent.com/plenoi/Clinic/master/ultima_all_clean.csv', sep=',', header = 0) #headerเป็นการบอกว่ามีหัวมั้ยจะมี0 1 กับ none
data_clean = df 
data_clean = data_clean.dropna(0,'any') #เมื่อเจอค่าว่างจะดรอปทั้งแถวและคอลัม
data_clean = data_clean.set_index('hn')
data_clean = data_clean.drop(['hiv','size','utmet','vgmet','surgery','pchemo'],1)
data_clean
```
